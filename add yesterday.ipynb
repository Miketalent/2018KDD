```r

# Load the data from the csv files
read.data = function(file = ''){
  ## Read the csv file
  air.quality <- read.csv(file, header = TRUE, 
                      stringsAsFactors = FALSE)
}
```

```r
bj.aq.grid = read.data('beijing_aq_grid_relation.csv')
ld.aq.grid = read.data('london_aq_grid_relation.csv')
# Create a list of the beijing station ID names
bj_names<-bj.aq.grid$station_id
# Manually input the station names that we need for London
ld_names <- ld.aq.grid$station_id
```

```r
bj_names
ld_names
```

```r

bj.fullStationData <- list()
ld.fullStationData <- list()
for (s in 1:35) {
    bj.fullStationData[[s]] <- read.data(paste("historicalWithTwoDaysAgo/",bj_names[s],"_historical.csv",sep=""))
    bj.fullStationData[[s]]$utc_time <- as.POSIXct(bj.fullStationData[[s]]$utc_time, origin="1970-01-01", format="%Y-%m-%d %H:%M:%S", tz="UTC")
}

for (s in 1:13) {
    ld.fullStationData[[s]] <- read.data(paste("historicalWithTwoDaysAgo/",ld_names[s],"_historical.csv",sep=""))
    ld.fullStationData[[s]]$utc_time <- as.POSIXct(ld.fullStationData[[s]]$utc_time, origin="1970-01-01", format="%Y-%m-%d %H:%M:%S", tz="UTC")
}
```

```r
head(ld.fullStationData[[1]],48)
```

```r
# Functions to find yesterday's values for beijing and london
bj_find_yesterday_val <- function(station_index, datetime, pollutant) {
    yesterdaytime = datetime - 86400 * 2
    lastVal <- bj.fullStationData[[station_index]][[pollutant]][bj.fullStationData[[station_index]]$utc_time==yesterdaytime][1]
    return (ifelse(is.na(lastVal),0,lastVal))
}

ld_find_yesterday_val <- function(station_index, datetime, pollutant) {
    yesterdaytime = datetime - 86400 * 2
    lastVal <- ld.fullStationData[[station_index]][[pollutant]][ld.fullStationData[[station_index]]$utc_time==yesterdaytime][1]
    return (ifelse(is.na(lastVal),0,lastVal))
}
```

```r
for (s in 1:13) {
    ld.fullStationData[[s]] <- ld.fullStationData[[s]][49:nrow(ld.fullStationData[[s]]),]
    
    filename <- paste("historicalWithLatest/",ld.fullStationData[[s]]$stationId[1],"_historical.csv",sep="")
    write.csv(ld.fullStationData[[s]], file = filename, row.names=FALSE)
}
```

```r
for (s in 1:35) {
    for (r in 1:nrow(bj.fullStationData[[s]])) {
        t <- bj.fullStationData[[s]]$utc_time[r]
        bj.fullStationData[[s]]$tdPM2.5[r] <- bj_find_yesterday_val(s,t,"PM2.5")
        bj.fullStationData[[s]]$tdPM10[r] <- bj_find_yesterday_val(s,t,"PM10")
        bj.fullStationData[[s]]$tdO3[r] <- bj_find_yesterday_val(s,t,"O3")   
    }
    
    filename <- paste("historicalWithTwoDaysAgo/",bj.fullStationData[[s]]$stationId[1],"_historical.csv",sep="")
    write.csv(bj.fullStationData[[s]], file = filename, row.names=FALSE)
}
```

```r
for (s in 1:13) {
    for (r in 1:nrow(ld.fullStationData[[s]])) {
        t <- ld.fullStationData[[s]]$utc_time[r]
        ld.fullStationData[[s]]$tdPM2.5[r] <- ld_find_yesterday_val(s,t,"PM2.5")
        ld.fullStationData[[s]]$tdPM10[r] <- ld_find_yesterday_val(s,t,"PM10")
    }
    
    filename <- paste("historicalWithTwoDaysAgo/",ld.fullStationData[[s]]$stationId[1],"_historical.csv",sep="")
    write.csv(ld.fullStationData[[s]], file = filename, row.names=FALSE)
}
```
